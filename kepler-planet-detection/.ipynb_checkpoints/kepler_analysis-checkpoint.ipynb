```python
import lightkurve as lk
import matplotlib.pyplot as plt

# Kepler ID'si (rnein, 6922244) olan bir yldzn verilerini indirin
kepler_id = '6922244'
search_result = lk.search_lightcurvefile(f'KIC {kepler_id}', mission='Kepler')
print(search_result)

lc = search_result.download().PDCSAP_FLUX
print(lc)

# Ik erisini iz
lc.plot()
plt.title(f'Kepler Light Curve for KIC {kepler_id}')
plt.show()
```

```python
# Trend karma ilemi
lc_detrended = lc.flatten()
lc_detrended.plot()
plt.title(f'Detrended Kepler Light Curve for KIC {kepler_id}')
plt.show()
```

```python
# Transit aramas yap
periodogram = lc_detrended.to_periodogram()
periodogram.plot()
plt.title(f'Periodogram for KIC {kepler_id}')
plt.show()

# En gl sinyali gster
best_period = periodogram.period_at_max_power
print(f'Best period: {best_period}')
```
